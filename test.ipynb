```python
import warnings
from agent import DQNAgent
from wrappers import make_env
from Policy import *
warnings.filterwarnings("ignore")


env = make_env('SuperMarioBros-1-1-v0')
state_dim = env.observation_space.shape
save_directory = "mario_dqn_model"
load_checkpoint = "checkpoint.pth"
```

```python
dqn_agent = DQNAgent(state_dim, env.action_space.n, EGreedyStategy(), save_directory)
dqn_result = dqn_agent.train(env, 0.9, "DQN", 500, 1, load_checkpoint)
```

```python
dqn_agent_softmax = DQNAgent(state_dim, env.action_space.n, SoftMaxStrategy(), save_directory)
dqn_softmax_result = dqn_agent_softmax.train(env, 0.9, "DQN", 500, 1, load_checkpoint)
```

```python
dqn_agent_EGreedyExp = DQNAgent(state_dim, env.action_space.n, EGreedyExpStrategy(), save_directory)
dqn_EGreedyExp_result = dqn_agent_EGreedyExp.train(env, 0.9, "DQN", 500, 1, load_checkpoint)
```

```python
dqn_agent_EGreedyLinear = DQNAgent(state_dim, env.action_space.n, EGreedyLinearStrategy(), save_directory)
dqn_EGreedyLinear_result = dqn_agent_EGreedyLinear.train(env, 0.9, "DQN", 500, 1, load_checkpoint)
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import os
```

```python
dqn_results=[dqn_result]
dqn_results = np.array(dqn_results)

dqn_softmax_results=[dqn_softmax_result]
dqn_softmax_results = np.array(dqn_softmax_results)

dqn_EGreedyExp_results=[dqn_EGreedyExp_result]
dqn_EGreedyExp_results = np.array(dqn_EGreedyExp_results)

dqn_EGreedyLinear_results=[dqn_EGreedyLinear_result]
dqn_EGreedyLinear_results = np.array(dqn_EGreedyLinear_results)
```

```python
dqn_mean_t, dqn_mean_r, dqn_mean_time = np.mean(dqn_results, axis=0).T
softmax_mean_t, softmax_mean_r, softmax_mean_time = np.mean(dqn_softmax_results, axis=0).T
Exp_mean_t, Exp_mean_r, Exp_mean_time = np.mean(dqn_EGreedyExp_results, axis=0).T
Linear_mean_t, Linear_mean_r, Linear_mean_time = np.mean(dqn_EGreedyLinear_results, axis=0).T

fig, axs = plt.subplots(2, 1, figsize=(15, 30), sharey=False, sharex=True)
axs[0].plot(dqn_mean_r, 'y', label='Epsilon Greedy', linewidth=2)
axs[0].plot(softmax_mean_r, 'b', label='Softmax', linewidth=2)
axs[0].plot(Exp_mean_r, 'r', label='Exponential Epsilon Greedy', linewidth=2)
axs[0].plot(Linear_mean_r, 'g', label='Linear Epsilon Greedy', linewidth=2)

axs[0].set_title('Moving Avg Reward (Training)')
axs[0].legend(loc='upper left')

axs[1].plot(dqn_mean_time, 'y', linewidth=2)
axs[1].plot(softmax_mean_time, 'b', linewidth=2)
axs[1].plot(Exp_mean_time, 'r', linewidth=2)
axs[1].plot(Linear_mean_time, 'g', linewidth=2)
axs[1].set_title('Training time')

plt.xlabel('Episodes')
fig.savefig("episode_rewards_plot3.png")
plt.show()
```
